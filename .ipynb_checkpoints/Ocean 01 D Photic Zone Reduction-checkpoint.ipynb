````python
def filter(v, ftype='hat', control1=3):
    """Operate on an XArray data array (with some checks) to produce a filtered version"""
    # pre-checks
    if not v.dims[0] == 'time': return v

    if ftype == 'hat': 
        n_passes = control1        # should be a kwarg
        len_v = len(v)
        for n in range(n_passes):
            source_data = np.copy(v) if n == 0 else np.copy(smooth_data)
            smooth_data = [source_data[i] if i == 0 or i == len_v - 1 else \
                0.5 * source_data[i] + 0.25 * (source_data[i-1] + source_data[i + 1]) \
                for i in range(len_v)]
        return smooth_data
    return v

def pull_data_array_by_name(ds, sensor_name):
    """completely extraneous (Delete Me) function"""
    return ds['sensor_name']

def dydx(y, x):
    """
    This is a simple 'centered' derivative: At location n it uses values at (n-1) and (n+1).
    Endpoints are hardcoded to zero so to do make them proper estimates.
    The sign +- also to do should be handled properly; varies with direction of profile up / down
    """ 
    len_y = len(y)
    return [0. if i == 0 or i == len_y - 1 else -(y[i+1] - y[i-1])/(x[i+1] - x[i-1]) for i in range(len_y)]

def ProfileCrawler(ds, verbose = False):
    """
    Walk a passed dataset to pin profiler start / stop times based on pressure
    """

    s = ds.seawater_pressure.to_series()         # pandas series, just pressure (and times)
    len_s = len(s)

    ascent_starts  = np.zeros(len_s)
    descent_starts = np.zeros(len_s)
    rest_starts    = np.zeros(len_s)
    
    threshold, nA0, nD0, nR0 = 1., 0, 0, 0

    for i in range(1, len_s - 5):         # 6 minute window
        
        # catch ascent
        if s[i-1] - s[i]   <= threshold and  \
           s[i]   - s[i+1] >= threshold and  \
           s[i+1] - s[i+2] >= threshold and  \
           s[i+2] - s[i+3] >= threshold and  \
           s[i+3] - s[i+4] >= threshold and  \
           s[i+4] - s[i+5] >= threshold:
            ascent_starts[i] = True
            nA0 += 1

        # catch descent
        if s[i-1] - s[i]   >= threshold and  \
           s[i]   - s[i+1] <= threshold and  \
           s[i+1] - s[i+2] <= threshold and  \
           s[i+2] - s[i+3] <= threshold and  \
           s[i+3] - s[i+4] <= threshold and  \
           s[i+4] - s[i+5] <= threshold:
            descent_starts[i] = True
            nD0 += 1
        
        # this variant is a little too liberal; false positives ~25%
        # why? Because twice daily there are stops on the way down for pH
        # catch rest
        if i >= 5 and \
           s[i-5] - s[i-4] <= -threshold and  \
           s[i-4] - s[i-3] <= -threshold and  \
           s[i-3] - s[i-2] <= -threshold and  \
           s[i-2] - s[i-1] <= -threshold and  \
           s[i-1] - s[i]   <= -threshold and  \
           s[i]   - s[i+1] >= -threshold:
            rest_starts[i] = True
            nR0 += 1
            
    if verbose: print(nA0, nD0, nR0)

    # These are index-accurate float-bool numpy arrays
    return ascent_starts, descent_starts, rest_starts
````

````python
# ds_CTD is a Dataset with 44640 values, one per minute for January 2019
# ds_CTD.data_vars -> seawater_temperature, seawater_pressure, practical_salinity, corrected_dissolved_oxygen, density
# There should be 279 profiles; but 278 were found. The last profile of the month did not go.

# This file read is recapitulated below to keep this code independent of that code
osb_ctd_nc_file = "./data/rca/ctd/osb_ctd_jan2019_1min.nc"
ds_CTD = xr.open_dataset(osb_ctd_nc_file)

ascent_starts, descent_starts, rest_starts = ProfileCrawler(ds_CTD)

ascent0, ascent1, descent0, descent1, rest0, rest1 = [], [], [], [], [], []

for i in range(len(ascent_starts)): 
    if ascent_starts[i]:  ascent0.append(i)
    if descent_starts[i]: descent0.append(i)
        
        
# above is reliable for Jan 2019
# rest0[] indices must disambiguate descent stops on midn/noon profiles favoring the last one
counter = 0       # will tracks profile index
                  # rest_starts[] is minute by minute with ones at possible rest starts
                  # descent0 already filled in with indices acting as "counter+1" stop i-indices 
                 
for i in range(len(ascent_starts)):   # one minute at a time
    if rest_starts[i] and i > descent0[counter]:    # this is a candidate i-value for rest0[]
                                                    # it assumes the first rest follows the first descent
        found_a_later_candidate = False
        j_loop_upper = descent0[counter + 1] if counter < len(ascent0) - 1 else len(ascent_starts)
        for j in range(i+1, j_loop_upper):
            if rest_starts[j]: found_a_later_candidate = True
        if not found_a_later_candidate:
            rest0.append(i)
            counter += 1
        
ascent1  = descent0.copy()
descent1 = rest0.copy()
rest1    = ascent0[1:].copy()
rest1.append(len(ascent_starts)-1)

# Use this to check the ordering
# for i in range(10): print(ascent0[i], ascent1[i], descent0[i], descent1[i], rest0[i], rest1[i])

# use this to check order logic for the full series
for i in range(len(ascent0)):
    if ascent0[i] >= descent0[i]: print('err 1', i)
    if descent0[i] >= rest0[i]: print('err 2', i)
    if ascent1[i] >= descent1[i]: print('err 3', i)
    if descent1[i] >= rest1[i]: print('err 4', i)

# at this point ideally all six index lists are the same length
print(len(ascent0), len(descent0), len(rest0), len(ascent1), len(descent1), len(rest1))

# the last values show rest1 ends at the last index of the source data 
# print(ascent0[-1], ascent1[-1])
# print(descent0[-1], descent1[-1])
# print(rest0[-1], rest1[-1])
````

````python
delta_ascent  = [ascent1[i] - ascent0[i] + 1 for i in range(len(ascent0))]
delta_descent = [descent1[i] - descent0[i] + 1 for i in range(len(descent0))]
delta_rest    = [rest1[i] - rest0[i] + 1 for i in range(len(rest0))]

print('Means, standard deviations for profile phases (minutes):')
print('  Ascent time:', round(np.mean(delta_ascent), 2), round(np.std(delta_ascent), 2))
print('  Descent time:', round(np.mean(delta_descent), 2), round(np.std(delta_descent), 2))
print('  Rest time:', round(np.mean(delta_rest[0:-1]), 2), round(np.std(delta_rest[0:-1]), 2))
print()
print('(Recall that two profiles of nine each day have slower, staged descents)')
print()
# print(delta_rest[-1])
````

````python
def ProfileChoicesChart(ctd, fluor, t, nf = 1):
    """
    Plot comparative charts for two profiles: CTD x 3, fluorescence x 3; left and
    right; based on time ranges t[0][0:1] and t[1][0:1]; using the 'hat' filter with param nf
    """
    
    # sanity check: print(t0, t1, type(t0), type(ctd.time[i0].data))
    
    ctd_slice   = [ctd.sel(time=slice(t[0][0], t[0][1])),   ctd.sel(time=slice(t[1][0], t[1][1]))]
    fluor_slice = [fluor.sel(time=slice(t[0][0], t[0][1])), fluor.sel(time=slice(t[1][0], t[1][1]))]

    ctd_t = [ctd_slice[0]['seawater_temperature'],       ctd_slice[1]['seawater_temperature']      ]   
    ctd_s = [ctd_slice[0]['practical_salinity'],         ctd_slice[1]['practical_salinity']        ]
    ctd_o = [ctd_slice[0]['corrected_dissolved_oxygen'], ctd_slice[1]['corrected_dissolved_oxygen']]
    
    fluor_a = [fluor_slice[0]['fluorometric_chlorophyll_a'],          fluor_slice[1]['fluorometric_chlorophyll_a']]
    fluor_c = [fluor_slice[0]['fluorometric_cdom'],                   fluor_slice[1]['fluorometric_cdom']]
    fluor_s = [fluor_slice[0]['total_volume_scattering_coefficient'], fluor_slice[1]['total_volume_scattering_coefficient']]
    
    ctd_t_filt = [filter(ctd_t[0], 'hat', nf), filter(ctd_t[1], 'hat', nf)]
    ctd_s_filt = [filter(ctd_s[0], 'hat', nf), filter(ctd_s[1], 'hat', nf)]
    ctd_o_filt = [filter(ctd_o[0], 'hat', nf), filter(ctd_o[1], 'hat', nf)]
               
    fluor_a_filt = [filter(fluor_a[0], 'hat', nf), filter(fluor_a[1], 'hat', nf)]
    fluor_c_filt = [filter(fluor_c[0], 'hat', nf), filter(fluor_c[1], 'hat', nf)]
    fluor_s_filt = [filter(fluor_s[0], 'hat', nf), filter(fluor_s[1], 'hat', nf)]
               
    ctd_d   = [ctd_slice[0]['seawater_pressure'], ctd_slice[1]['seawater_pressure']]
    fluor_d = [fluor_slice[0]['int_ctd_pressure'], fluor_slice[1]['int_ctd_pressure']]

    temperature_lbound, temperature_ubound           =   8.,        12.
    salinity_lbound, salinity_ubound                 =  32.,        34.
    dissolved_oxygen_lbound, dissolved_oxygen_ubound = 100.,       300.
    chlora_lbound, chlora_ubound                     =  -1.,         2.
    cdom_lbound, cdom_ubound                         =   1.2,        2.2
    scatt_lbound, scatt_ubound                       =   0.000075,    .00035

    fig, axs = plt.subplots(6,2,figsize=(16, 15), tight_layout=True)

    axs[0][0].plot(ctd_t_filt[0], ctd_d[0], marker='o', markersize = 6., color='k', markerfacecolor='c')
    axs[0][0].set(xlim = (temperature_lbound, temperature_ubound), ylim = (200., 0.), title='temperature')
    axs[1][0].plot(ctd_s_filt[0], ctd_d[0], marker='s', markersize = 6., color='g', markerfacecolor='m')
    axs[1][0].set(xlim = (salinity_lbound, salinity_ubound), ylim = (200., 0.), title='salinity')
    axs[2][0].plot(ctd_o_filt[0], ctd_d[0], marker='D', markersize = 6., color='b', markerfacecolor='y')
    axs[2][0].set(xlim = (dissolved_oxygen_lbound, dissolved_oxygen_ubound), ylim = (200., 0.), title='dissolved oxygen')
    axs[3][0].plot(fluor_a_filt[0], fluor_d[0], marker='o', markersize = 6., color='k', markerfacecolor='c')
    axs[3][0].set(xlim = (chlora_lbound, chlora_ubound), ylim = (200., 0.), title='chlorophyll a')
    axs[4][0].plot(fluor_c_filt[0], fluor_d[0], marker='s', markersize = 6., color='g', markerfacecolor='m')
    axs[4][0].set(xlim = (cdom_lbound, cdom_ubound), ylim = (200., 0.), title='cdom')
    axs[5][0].plot(fluor_s_filt[0], fluor_d[0], marker='D', markersize = 6., color='b', markerfacecolor='y')
    axs[5][0].set(xlim = (scatt_lbound, scatt_ubound), ylim = (200., 0.), title='total volume backscatter')

    axs[0][1].plot(ctd_t_filt[1], ctd_d[1], marker='o', markersize = 6., color='k', markerfacecolor='c')
    axs[0][1].set(xlim = (temperature_lbound, temperature_ubound), ylim = (200., 0.), title='temperature')
    axs[1][1].plot(ctd_s_filt[1], ctd_d[1], marker='s', markersize = 6., color='g', markerfacecolor='m')
    axs[1][1].set(xlim = (salinity_lbound, salinity_ubound), ylim = (200., 0.), title='salinity')
    axs[2][1].plot(ctd_o_filt[1], ctd_d[1], marker='D', markersize = 6., color='b', markerfacecolor='y')
    axs[2][1].set(xlim = (dissolved_oxygen_lbound, dissolved_oxygen_ubound), ylim = (200., 0.), title='dissolved oxygen')
    axs[3][1].plot(fluor_a_filt[1], fluor_d[1], marker='o', markersize = 6., color='k', markerfacecolor='c')
    axs[3][1].set(xlim = (chlora_lbound, chlora_ubound), ylim = (200., 0.), title='chlorophyll a')
    axs[4][1].plot(fluor_c_filt[1], fluor_d[1], marker='s', markersize = 6., color='g', markerfacecolor='m')
    axs[4][1].set(xlim = (cdom_lbound, cdom_ubound), ylim = (200., 0.), title='cdom')
    axs[5][1].plot(fluor_s_filt[1], fluor_d[1], marker='D', markersize = 6., color='b', markerfacecolor='y')
    axs[5][1].set(xlim = (scatt_lbound, scatt_ubound), ylim = (200., 0.), title='total volume backscatter')
    
    print('left profile', str(t[0][0])[:16], 'to', str(t[0][1])[:16], \
          '... right profile', str(t[1][0])[:16], 'to', str(t[0][1])[:16])
    return

left0, right0 = 0, 1

t00 = dt64(ds_CTD.time[ascent0[left0]].data)   
t01 = dt64(ds_CTD.time[ascent1[left0]].data)   
t10 = dt64(ds_CTD.time[ascent0[right0]].data)   
t11 = dt64(ds_CTD.time[ascent1[right0]].data) 

nProfiles = len(ascent0)

def SliderRelay(nLeft, nRight):
    global t00, t01, t10, t11, ds_CTD, ds_Fluorometer  
    t00 = dt64(ds_CTD.time[ascent0[nLeft]].data)
    t01 = dt64(ds_CTD.time[ascent1[nLeft]].data)     
    t10 = dt64(ds_CTD.time[ascent0[nRight]].data)
    t11 = dt64(ds_CTD.time[ascent1[nRight]].data)
    ProfileChoicesChart(ds_CTD, ds_Fluorometer, [[t00,t01], [t10,t11]], 1)


interact(SliderRelay, nLeft=widgets.IntSlider(min=0, max=nProfiles-1, step=1, value=left0, \
                                          continuous_update=False, description='left'), \
                      nRight=widgets.IntSlider(min=0, max=nProfiles-1, step=1, value=right0, \
                                       continuous_update=False, description='right'))
               
````

## Profile as a class

The idea here is to create a Python class 'Profile' that can be used to organize many profiles.

````python
class Profile:
    """A water column profile"""
    
    def __init__(self, t0='2019-01-01T00:26:05', t1='2019-01-01T01:37:55', d0=191.268063, d1=6.618323, site='osb'):
        self.t0 = dt64(t0)
        self.t1 = dt64(t1)
        self.d0 = d0
        self.d1 = d1
        self.site = site

    def readout(self):
        print("Profile start", self.t0, 'duration', self.t1 - self.t0)

p = Profile()
p.readout()
````
